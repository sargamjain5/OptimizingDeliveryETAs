```python
import pandas as pd

df = pd.read_csv("../data/delivery_data.csv")
```

```python
df["delay_ratio"] = df["actual_time"] / df["osrm_time"]
```

```python
df["segment_delay_ratio"] = (
    df["segment_actual_time"] /
    df["segment_osrm_time"]
)
```
